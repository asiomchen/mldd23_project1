```python
from bayes_opt import BayesianOptimization
from src.disc.discriminator import Discriminator
import torch
import random

latent_size = 64
model = Discriminator(latent_size=latent_size, use_sigmoid=False)
model.load_state_dict(torch.load('models/discr_d2_mandarynka_epoch_100/epoch_150.pt'))

args = {str(n): 0.0 for n in range(latent_size)}
pbounds = {str(p):(-1,1) for p in range(latent_size)}

def foo(**args):
    dict = {**args}
    input_tensor = torch.tensor(list(dict.values()))
    input_tensor = input_tensor.to(torch.float32)
    pred = model(input_tensor)
    output = pred.cpu().detach().numpy()[0]
    print(output)
    return output
```

```python
optimizer = BayesianOptimization(
    f=foo,
    pbounds=pbounds,
    random_state=random.randint(0, 100000),
)

optimizer.maximize(
    init_points=100,
    n_iter=20,
)
```

```python
print(optimizer.max)
```
